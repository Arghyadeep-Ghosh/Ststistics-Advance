```python
''' Question1: Define the z-statistic and explain its relationship to the standard normal distribution. How is the z-statistic used in hypothesis testing?

Ans) Z-Statistic and its Relationship to the Standard Normal Distribution

The Z-statistic (or Z-score) is a measurement that describes how many
standard deviations a data point (or sample mean) is away from the population mean.
It is useful to standardize data points, enabling comparisons across different datasets.

The formula for the Z-statistic is:
    Z = (X - μ) / σ
where:
- X = observed value (or sample mean)
- μ = population mean
- σ = population standard deviation

The Z-statistic follows the standard normal distribution, which has:
- Mean (μ) = 0
- Standard deviation (σ) = 1

A Z-score of:
- 0 means the value is at the mean.
- Positive indicates the value is above the mean.
- Negative indicates the value is below the mean.


 Use of Z-Statistic in Hypothesis Testing


The Z-statistic is often used for hypothesis testing, especially when:
1. The population variance is known or
2. The sample size is large (n > 30).

Steps in Z-test Hypothesis Testing:
1. State the hypotheses:
   - Null hypothesis (H₀): No effect or no difference.
   - Alternative hypothesis (H₁): There is an effect or difference.

2. Calculate the Z-statistic using the formula relevant to the test type.

3. Find the critical Z-value (Z-critical) from the Z-table for the chosen
   significance level (e.g., α = 0.05).

4. Compare the Z-statistic with Z-critical:
   - If Z-statistic falls in the rejection region (extreme tail), reject H₀.
   - If it is within the acceptance region, fail to reject H₀.

Example:
- For a one-tailed test with α = 0.05, reject H₀ if Z > 1.645.
- For a two-tailed test, reject H₀ if Z < -1.96 or Z > 1.96.

The Z-statistic makes it easier to compare test results using the standard normal
distribution and is useful for large-sample or known-variance cases.
'''


```

```python
'''
Question2 : What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is
very small (e.g., 0.01)?

Ans)  The p-value is the probability of obtaining test results at least as extreme as the observed results,
      under the assumption that the null hypothesis (H₀) is true. It helps us measure the strength of evidence
      against the null hypothesis.

In hypothesis testing, the p-value is used to decide whether to reject H₀:
- If the p-value is **small** (less than the significance level, α), it suggests strong evidence against H₀,
  leading to its rejection.
- If the p-value is **large**, it indicates that the observed result could happen due to chance, and we fail to reject H₀.

The most commonly used significance levels are:
- α = 0.05 (5%)
- α = 0.01 (1%)


Meaning of a Very Small p-value (e.g., 0.01)


If the p-value is very small (e.g., 0.01), it means that the observed data is very unlikely under the assumption
that the null hypothesis is true. In this case:
- If α = 0.05, the p-value (0.01) is smaller than α, so we reject the null hypothesis.
- The smaller the p-value, the stronger the evidence against H₀.

A p-value of 0.01 indicates that there is only a 1% chance of obtaining such extreme results if H₀ is true.
Thus, we can conclude that the observed result is statistically significant at the 5% level (α = 0.05) or
even the 1% level (α = 0.01).


Summary: p-value in Hypothesis Testing


1. **Small p-value** (< α): Strong evidence against H₀ → Reject H₀.
2. **Large p-value** (≥ α): Weak evidence against H₀ → Fail to reject H₀.

The p-value provides a flexible way to interpret test results without being bound to a fixed threshold
like α. Smaller p-values indicate stronger evidence against the null hypothesis.
'''
```

```python
'''
Question3: Compare and contrast the binomial and Bernoulli distributions.

Ans)  Comparison of Binomial and Bernoulli Distributions

Both the Binomial and Bernoulli distributions are used to model binary outcomes,
but they differ in scope and application. Below is a comparison:

1. **Definition**:
   - **Bernoulli Distribution**: Models a single trial with two outcomes:
     success (1) or failure (0).
   - **Binomial Distribution**: Models the total number of successes
     in a fixed number of independent Bernoulli trials.

2. **Random Variable**:
   - **Bernoulli**: A single random variable with possible outcomes {0, 1}.
   - **Binomial**: A random variable representing the total number of successes
     across multiple independent Bernoulli trials.

3. **Number of Trials**:
   - **Bernoulli**: Only 1 trial.
   - **Binomial**: n independent trials, where n ≥ 1.

4. **Probability of Success (p)**:
   - **Bernoulli**: A single probability p for one trial.
   - **Binomial**: The same probability p is used across all n trials.

5. **Probability Mass Function (PMF)**:
   - **Bernoulli**:
     P(X = x) = p^x * (1 - p)^(1 - x), for x ∈ {0, 1}
   - **Binomial**:
     P(X = k) = C(n, k) * p^k * (1 - p)^(n - k), for k ∈ {0, 1, ..., n}
     where C(n, k) = n! / (k! * (n - k)!)

6. **Mean and Variance**:
   - **Bernoulli**:
     - Mean: μ = p
     - Variance: σ² = p(1 - p)
   - **Binomial**:
     - Mean: μ = n * p
     - Variance: σ² = n * p * (1 - p)

7. **Use Case**:
   - **Bernoulli**: Used to model a single event with two possible outcomes
     (e.g., flipping a coin once).
   - **Binomial**: Used to model the number of successes over multiple trials
     (e.g., flipping a coin 10 times and counting the heads).

8. **Relationship**:
   - The **Binomial distribution** is a generalization of the **Bernoulli distribution**.
     If n = 1 in a Binomial distribution, it becomes a Bernoulli distribution.

Summary:
- **Bernoulli**: One trial, two outcomes.
- **Binomial**: Multiple independent trials, counting the number of successes.
Both distributions are fundamental in probability theory and statistics, with Binomial
being used to model multiple Bernoulli events.
'''
```

```python
# Question 4: Under what conditions is the binomial distribution used, and how does it relate to the Bernoulli distribution?
# Ans)
# Conditions for Using the Binomial Distribution and Its Relation to the Bernoulli Distribution

'''
## Conditions for Using the Binomial Distribution
The Binomial distribution is appropriate under the following conditions:

1. **Fixed Number of Trials (n)**:
   - The experiment consists of a specific number of independent trials.

2. **Binary Outcomes**:
   - Each trial has only two possible outcomes: success (1) or failure (0).

3. **Constant Probability (p)**:
   - The probability of success (p) remains the same for each trial.

4. **Independence of Trials**:
   - The outcome of one trial does not affect the outcome of other trials.

Examples:
- Flipping a coin 10 times and counting the heads.
- Rolling a die multiple times and checking for a particular number.

---

## Relation Between Binomial and Bernoulli Distributions
1. **Generalization**:
   - The **Binomial distribution** is a generalization of the **Bernoulli distribution**.
   - A **Bernoulli distribution** describes the outcome of a **single trial**.

2. **Multiple Bernoulli Trials**:
   - When n independent Bernoulli trials are repeated and we count the number of successes,
     the result follows a **Binomial distribution**.

3. **Special Case**:
   - If the number of trials (n) in a Binomial distribution is 1, it becomes a **Bernoulli distribution**.
     In this sense, a Bernoulli distribution is a special case of the Binomial distribution with n = 1.

---

## Summary
- **Binomial distribution**: Models the number of successes over multiple independent trials.
- **Bernoulli distribution**: Models a single trial with binary outcomes.
- If n = 1 in a Binomial distribution, it simplifies to a Bernoulli distribution.
Both distributions are used when dealing with events that have two outcomes, but the Binomial distribution
is used when the experiment involves **multiple trials**.
'''
```

```python
'''
Question5: What are the key properties of the Poisson distribution, and when is it appropriate to use this
distribution?

Ans)
# Key Properties of the Poisson Distribution and When to Use It


## Key Properties of the Poisson Distribution

1. **Discrete Distribution**:
   - The Poisson distribution models the number of events occurring in a fixed interval
     (e.g., time, space, area) that are discrete (non-negative integers).

2. **Parameter (λ)**:
   - λ (lambda) represents both the **mean** and **variance** of the distribution.
   - λ is the expected number of events occurring in the given interval.

3. **Probability Mass Function (PMF)**:
   - The PMF of the Poisson distribution is:
     P(X = k) = (λ^k * e^(-λ)) / k!,  where k = 0, 1, 2, ...
   - Here, e is the Euler's number (approximately 2.718).

4. **Mean and Variance**:
   - Mean (μ) = λ
   - Variance (σ²) = λ

5. **Memoryless Property**:
   - The number of events in non-overlapping intervals is independent.
   - Knowing the number of events in one interval gives no information about other intervals.

6. **Skewness**:
   - The distribution is skewed to the right, but as λ increases, it becomes more symmetric.

---

## When is the Poisson Distribution Appropriate?

The Poisson distribution is used when:
1. **Counting Events in a Fixed Interval**:
   - The number of events (e.g., customer arrivals, phone calls) in a given period of time or space.

2. **Events Occur Independently**:
   - The occurrence of one event does not affect the probability of another event occurring.

3. **Low Probability but Large Number of Trials**:
   - Events are rare on a small scale (low probability), but over a long period or large sample size,
     some occurrences are expected.

4. **Constant Rate of Occurrence (λ)**:
   - The average rate of events remains constant over the given interval.

---

## Examples of Poisson Distribution Use
- Number of calls received by a call center per hour.
- Number of cars passing through a toll booth in a day.
- Number of emails received in an inbox over a period.

---

## Summary
- The Poisson distribution models the occurrence of events over a fixed interval.
- It assumes events are independent and occur at a constant rate (λ).
- It is especially useful when the events are rare but happen over a large scale or time period.
'''




```

```python
'''
Question6: Define the terms "probability distribution" and "probability density function" (PDF). How does a
PDF differ from a probability mass function (PMF)?

Ans) # Probability Distribution, Probability Density Function (PDF), and Difference from PMF

## 1. What is a Probability Distribution?
A **probability distribution** describes how the probabilities of all possible outcomes
of a random variable are distributed. It can be discrete or continuous, depending on
the nature of the variable.

- **Discrete Probability Distribution**: Deals with countable outcomes (e.g., rolling a die).
- **Continuous Probability Distribution**: Deals with uncountable outcomes, often represented
  over an interval (e.g., height of people).

The distribution summarizes the likelihood of different outcomes occurring.

---

## 2. What is a Probability Density Function (PDF)?
A **Probability Density Function (PDF)** describes the probability distribution of a **continuous**
random variable. It gives the likelihood of the random variable falling within a particular range.

- **Properties of a PDF**:
  1. The value of the PDF is non-negative for all outcomes.
  2. The area under the curve of the PDF over the entire range of possible outcomes equals **1**.
  3. For continuous variables, the probability of a specific value is **0**; we only consider probabilities
     over intervals.

Example: The PDF of a normal distribution describes how the data points are distributed around the mean.

---

## 3. How does a PDF Differ from a PMF?
- **PDF (Probability Density Function)**:
  - Used for **continuous** random variables.
  - Represents the likelihood over an interval, not individual points.
  - Probability is calculated as the **area under the curve** over the desired range.

- **PMF (Probability Mass Function)**:
  - Used for **discrete** random variables.
  - Assigns a specific probability to each possible outcome.

---

## 4. Summary of Differences Between PDF and PMF
- **Random Variable**:
  - PDF: Continuous.
  - PMF: Discrete.

- **Probability Calculation**:
  - PDF: Probability over an interval (area under the curve).
  - PMF: Probability at specific points (sum of individual probabilities).

- **Value at a Point**:
  - PDF: Probability of a single point is 0.
  - PMF: Each point has a non-zero probability.

---

## Example
- **PDF**: Probability that a person’s height is between 160 cm and 170 cm.
- **PMF**: Probability of rolling a 4 on a 6-sided die.
'''
```

```python
'''
Question7: Explain the Central Limit Theorem (CLT) with example.

Ans) # Central Limit Theorem (CLT) with Example

## 1. What is the Central Limit Theorem (CLT)?
The **Central Limit Theorem (CLT)** states that the sampling distribution of the **sample mean**
approaches a normal distribution as the sample size increases, **regardless of the original distribution**
of the population.

This means that if you take many random samples from a population with any distribution
(e.g., uniform, skewed, or normal) and calculate their means, the distribution of these sample means
will be approximately normal, given a large enough sample size.

---

## 2. Key Points of the CLT:
1. **Sample Size (n)**:
   - As the sample size increases, the sampling distribution of the mean becomes more like a normal distribution.
   - A sample size of **n ≥ 30** is generally considered sufficient for the CLT to hold.

2. **Mean and Standard Deviation**:
   - Mean of the sampling distribution = Population mean (μ).
   - Standard deviation of the sampling distribution (also called **Standard Error**) = σ / √n,
     where:
     - σ = Population standard deviation
     - n = Sample size

3. **Independence**:
   - The samples must be independent of each other.

---

## 3. Why is the CLT Important?
The CLT is fundamental to statistics because it allows us to:
- Use **normal distribution approximations** to perform hypothesis tests, even when the population
  is not normally distributed.
- Make **inferences about population parameters** using sample statistics.

---

## 4. Example of the CLT:
Imagine we are measuring the heights of students in a school. The population distribution of students' heights
might be skewed (e.g., more students are of average height, with fewer very tall or very short students).

- **Step 1**: Take multiple random samples of size 30 from the population.
- **Step 2**: Calculate the mean height for each sample.
- **Step 3**: Plot the distribution of these sample means.

Result:
- As the number of samples increases, the distribution of the sample means will look more like a **normal distribution**,
  even if the original population distribution was skewed.

---

## 5. Summary:
- The CLT states that the distribution of sample means approaches normality as sample size increases.
- It allows us to use normal distribution-based techniques for analysis, even when the population is not normal.
- Larger sample sizes lead to more accurate approximations.

The CLT is the foundation for many statistical procedures, such as **confidence intervals** and **hypothesis testing**.
'''
```

```python
'''Question9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample
size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to
reject the null hypothesis?

 Task: Write Python code to calculate the z-score and p-value for the given data.

Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing

# Importing necessary library
from scipy import stats
'''
#Ans)
# Given data
sample_mean = 105  # Mean of the sample
population_mean = 100  # Mean of the population
std_dev = 15  # Standard deviation of the population
sample_size = 25  # Number of observations in the sample
significance_level = 0.05  # Significance level (alpha)

# Calculating the standard error of the mean (SEM)
# Formula: SEM = std_dev / sqrt(sample_size)
standard_error = std_dev / (sample_size ** 0.5)

# Calculating the z-score
# Formula: z = (sample_mean - population_mean) / standard_error
z_score = (sample_mean - population_mean) / standard_error

# Calculating the p-value for the two-tailed test
# Using the survival function (sf) to get the p-value from z-score
p_value = 2 * stats.norm.sf(abs(z_score))  # Two-tailed p-value

# Decision: Check if we reject or fail to reject the null hypothesis
# If p-value < significance_level, we reject the null hypothesis
if p_value < significance_level:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Printing the results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(f"Decision: {decision}")
```

```python
'''Question10: Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python.
Generate 1,000 samples and plot the distribution. What is the expected mean and variance?

Task: Use Python to generate the data, plot the distribution, and calculate the mean and variance.

Objective: Understand the properties of a binomial distribution and verify them through simulation  '''

#Ans)
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Given parameters for the binomial distribution
n_trials = 10  # Number of trials
p_success = 0.6  # Probability of success
num_samples = 1000  # Number of samples to generate

# Simulating the binomial distribution
samples = np.random.binomial(n=n_trials, p=p_success, size=num_samples)

# Calculating the expected mean and variance
# Mean = n * p
expected_mean = n_trials * p_success

# Variance = n * p * (1 - p)
expected_variance = n_trials * p_success * (1 - p_success)

# Calculating the mean and variance of the generated samples
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

# Plotting the distribution of the samples
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=range(0, n_trials + 2), edgecolor='black', alpha=0.7, density=True)
plt.title('Binomial Distribution (n=10, p=0.6, 1000 samples)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(0, n_trials + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Printing the results
print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```
